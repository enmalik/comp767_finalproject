### scores

```python
def get_scores(mean_arr, window=50):
    rolling_means = []
    
    for i in range(len(mean_arr)-window):
        r_mean = np.mean(mean_arr[i:i+window])
        rolling_means.append(r_mean)
        
    max_idx = np.argmax(rolling_means)
    ep_conv = max_idx + 1
    max_rolling_score = rolling_means[max_idx]
    
    return ep_conv, max_rolling_score
```
